# -> Predicting Occupation wrt Age and Stress Level with KNN

```python
import plotly.express as px
fig = px.scatter(df2, x="Age", y="Stress Level",title="Age Vs Stress Level",
                 size="Stress Level", color="Stress Level",hover_data=["Occupation"])
fig.update_layout(title_font = dict(size = 50),title_font_color ="Honey dew",title_font_family = "Benett Sans Serif")
fig.update_layout(title_x = 0.47)
fig.show()
```

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.neighbors import KNeighborsClassifier
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap
```

# **#Taking Independent Variables as Age and Stress Level and dependent Variable as Occupation**

```python
X = df2[['Age', 'Stress Level']]
y = df2['Occupation']
```

# **#Encode the target variable so as to give unique numerical value to Occupation Categorical data**

```python
le = LabelEncoder()
y_encoded = le.fit_transform(y)
```

# **#Split the data into training and testing sets (Testing on 20% of the data)**

```python
X_train, X_test, y_train, y_test = train_test_split(X, y_encoded, test_size=0.2, random_state=42)
```

# **#Applying KNN by taking neigbours as 3 [It searches for 3 neigbours to make the predictions]**

```python
knn = KNeighborsClassifier(n_neighbors=3)
knn.fit(X_train, y_train)
```

```python
# Creating mesh grid
h = .02
x_min, x_max = X['Age'].min() - 1, X['Age'].max() + 1
y_min, y_max = X['Stress Level'].min() - 1, X['Stress Level'].max() + 1
xx, yy = np.meshgrid(np.arange(x_min, x_max, h), np.arange(y_min, y_max, h))
Z = knn.predict(np.c_[xx.ravel(), yy.ravel()]).reshape(xx.shape)

#Ensuring each class has a distinct color
plt.figure(figsize=(10, 6))
plt.pcolormesh(xx, yy, Z, cmap=ListedColormap(plt.cm.tab20.colors[:len(le.classes_)]))
scatter = plt.scatter(X['Age'], X['Stress Level'], c=y_encoded, cmap=ListedColormap(plt.cm.tab20.colors[:len(le.classes_)]), edgecolor='k', s=20)
plt.xlabel('Age')
plt.ylabel('Stress Level')
plt.title('KNN Classification (Occupation)')

# Creating legend for each class
handles = [plt.Line2D([0], [0], marker='o', color='w', markerfacecolor=plt.cm.tab20(i / len(le.classes_)), markersize=10) for i in range(len(le.classes_))]
plt.legend(handles, le.classes_, bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

# -> People with particular age and stress Levels belongs to particular Occupation.

# **ML - DECISION TREES**

# Predicting whether Disorder is there or not by splitting the nodes on the basis of Age and Stress Levels

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.tree import export_graphviz
import graphviz
import pydotplus
```

# **#Mapping the Sleep Disorders to 0 and 1**

# 0 -> No Disorder and 1 -> either Sleep Apnea or Insomnia

```python
d1 = {"None": 0, "Sleep Apnea": 1, "Insomnia": 1}
df['Sleep Disorder(0/1)'] = df['Sleep Disorder'].map(d1)
```

# **Taking Sleep Disorder as dependent Variable and Age, Sleep Hours as independent Variable.**

```python
x = df[["Age", "Sleep Hours"]]
y = df["Sleep Disorder(0/1)"]
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.25, random_state=42)
```

# **#Applying Decision Tree Technique and saving the graph to png**

```python
clf = DecisionTreeClassifier(random_state=42)
clf.fit(x_train, y_train)
class_names_actual = ["No Disorder", "Disorder"]
dot_data = export_graphviz(clf, out_file=None,
                           feature_names=["Age", "Sleep Hours"],
                           class_names=class_names_actual,
                           filled=True, rounded=True,
                           special_characters=True)
graph = pydotplus.graph_from_dot_data(dot_data)
```

```python
from IPython.display import Image
Image(graph.create_png())
```

 # -> True shows that the decision tree has been saved as png file.

# **#Prediciting y_train and y_test values from x_train and x_test**

```python
y_train_pred = clf.predict(x_train)
y_test_pred = clf.predict(x_test)
```

# **#confusion Matrix**

```python
from sklearn.metrics import confusion_matrix
```

```python
confusion_matrix(y_train, y_train_pred)
```

```python
cf_matrix = confusion_matrix(y_test, y_test_pred)
cf_matrix
```

# **#Visualizing confusion Matrix on Heat-Map**

```python
import seaborn as sns
sns.heatmap(cf_matrix, annot=True, cmap='Blues')
plt.xlabel('Predicted Values')
plt.ylabel('Actual Values')
plt.title('Confusion Matrix')
plt.show()
```

# -> Diagonal values show the correct predictions and off-diagonal values show the incorrect predicitions.
# -> It can be used to find the accuracy as well.

# **#finding Accuracy Manually**
# Correct predictions = 12+8 = 20
# total predicitions = 12+8+9+4 = 33
# So, Accuracy is 20/33 = 0.60 or 60%

# **#Determining Accuracy Score with sklearn**

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_train, y_train_pred)
```

```python
accuracy_score(y_test, y_test_pred)
```

# **ML - Random Forest Classifier**

# -> It applies multiple Decision Trees on the dataset and then takes the average to get the best fit model of Decision Tree for the dataset.

# **#Importing the required Libraries for Random Forest**

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.ensemble import RandomForestClassifier
import matplotlib.pyplot as plt
```

# **#Taking Different features of dataset and also converting the Categorical data of the dataset to numeric labels**

```python
le = LabelEncoder()
for col in ['Gender', 'Occupation', 'BMI Category', 'Sleep Disorder']:
    df[col+" Label Encoder"] = le.fit_transform(df[col])

features = ['Gender Label Encoder', 'Age', 'Occupation Label Encoder', 'Sleep Hours', 'Sleep Quality in Hours',
            'Physical Activity Level %', 'BMI Category Label Encoder',
            'Heart Rate', 'Daily Steps', 'Sleep Disorder Label Encoder']
X = df[features]
y = df["Stress Level"]
```

# **#Testing the model on 25% of the dataset**

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

# **#Taking the Best Decision Tree for Random Forest Model**
